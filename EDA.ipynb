# mean,variance,std-dev

```python
print("*******mean*******")  
print(np.mean(df_setosa["petal_length"]))
print(np.mean(df_verisicolor["petal_length"]))
print(np.mean(df_virginica["petal_length"]))
print(np.mean(np.append(df_setosa['petal_length'],50)))
# an outlier added to setosa petal length the mean drastically changes 
```

```python
print("*******varience******")
print(np.var(df_setosa["petal_length"]))
print(np.var(df_verisicolor["petal_length"]))
print(np.var(df_virginica["petal_length"]))
print(np.var(np.append(df_setosa['petal_length'],50)))
```

```python
print("*******standard-deviation******")
print(np.std(df_setosa["petal_length"]))
print(np.std(df_verisicolor["petal_length"]))
print(np.std(df_virginica["petal_length"]))
print(np.std(np.append(df_setosa['petal_length'],50)))
```

```python
print("*******median*******")  
print(np.median(df_setosa["petal_length"]))
print(np.median(df_verisicolor["petal_length"]))
print(np.median(df_virginica["petal_length"]))
print(np.median(np.append(df_setosa['petal_length'],50)))
# an outlier added to setosa petal length but the median remains same
print(np.median(np.append(df_setosa['petal_length'],50,67)))
# an outliers added to setosa petal length but the median remains same
```

```python
counts,bin_edges=np.histogram(df_setosa['petal_length'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)   #blue line is pdf
plt.plot(bin_edges[1:],cdf)#yellow line is cdf
plt.xlabel("petal_length")
plt.show()
```

```python
counts,bin_edges=np.histogram(df_setosa['sepal_length'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)   #blue line is pdf
plt.plot(bin_edges[1:],cdf)#yellow line is cdf
plt.xlabel("sepal_length")
plt.show()
```
